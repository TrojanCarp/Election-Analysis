```python
# For data
import pandas as pd
from pandas import Series,DataFrame
import numpy as np

# For visualization
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
plt.style.use('fivethirtyeight')

#Futuristic Division
from __future__ import division
```

```python
tip_df = sns.load_dataset('tips')
```

```python
tip_df.head()
```

```python
tip_df.pivot_table('tip',index = 'sex',aggfunc='mean')
```

```python
#Linear Regression plot for tip vs total bill
fig1=sns.lmplot('total_bill','tip',
           tip_df,scatter_kws={'marker':'o','color':'indianred'})
fig1.set(xlabel='Total Bill',ylabel='Tip',title='Tip Amount vs. Total Bill')
```

```python
#Polinomial fit
fig=sns.lmplot('total_bill','tip',
           tip_df,order=4,scatter_kws={'marker':'o','color':'indianred'})
fig.set(xlabel='Total Bill',ylabel='Tip',title='Tip Amount vs. Total Bill')
```

```python
tip_df.head()
```

```python
tip_df['Pct Tip'] = (tip_df.tip/tip_df.total_bill)*100
tip_df.head()
```

```python
tipsize = sns.lmplot('size','Pct Tip', tip_df, x_jitter= .1)
tipsize.set(xlabel='Size of Party', ylabel='Tip Percentage',title='Tip Percentage vs. Size of Party')
```

```python
gendertip = sns.lmplot('total_bill','Pct Tip',tip_df,hue='sex')
gendertip.set(xlabel='Total Bill', ylabel='Tip %', title = 'Tip % by Gender')
```

```python
tipbyday = sns.lmplot('total_bill','Pct Tip',tip_df,hue='day')
tipbyday.set(xlabel='Total Bill', ylabel='Tip %', title = 'Tip % by Day')
```
